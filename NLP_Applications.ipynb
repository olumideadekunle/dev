## 1. Using spaCy for Named Entity Recognition (NER)

````python
# Install spaCy: !pip install spacy
import spacy

# Load pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "Barack Obama was born on August 4, 1961, in Honolulu, Hawaii."

# Process text
doc = nlp(text)

# Extract entities
print("Entities and their labels:")
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
````

## 2. Using NLTK for Text Tokenization

````python
# Install NLTK: !pip install nltk
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added to explicitly download punkt_tab

# Input text
text = "Natural Language Processing is exciting! Let's learn it together."

# Tokenize text into words
tokens = nltk.word_tokenize(text)
print("Word Tokens:", tokens)

# Tokenize text into sentences
sentences = nltk.sent_tokenize(text)
print("Sentence Tokens:", sentences)
````

## 3. Using Hugging Face Transformers for Sentiment Analysis

````python
# Install Hugging Face Transformers: !pip install transformers
from transformers import pipeline

# Load pre-trained sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Input text
text = ["I love this product!", "I hate waiting in line."]

# Perform sentiment analysis
results = sentiment_pipeline(text)
for i, result in enumerate(results):
    print(f"Text: {text[i]}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}")
````

## 4. Using spaCy for Dependency Parsing

````python
# Dependency Parsing with spaCy
doc = nlp("The quick brown fox jumps over the lazy dog.")
print("Dependency Parsing:")
for token in doc:
    print(f"Token: {token.text}, Dependency: {token.dep_}, Head: {token.head.text}")
````

## 5. Using NLTK for Stemming and Lemmatization

````python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Example words
words = ["running", "runner", "ran", "easily", "flying"]

# Apply stemming and lemmatization
print("Stemming Results:")
for word in words:
    print(f"{word} -> {stemmer.stem(word)}")

print("\nLemmatization Results:")
for word in words:
    print(f"{word} -> {lemmatizer.lemmatize(word, pos=wordnet.VERB)}")
````

## 6. Using Hugging Face for Text Generation (GPT-2)

## 7. Using spaCy for Part-of-Speech Tagging

````python
# Part-of-Speech Tagging with spaCy
doc = nlp("Apple is looking at buying a U.K. startup.")
print("Part-of-Speech Tags:")
for token in doc:
    print(f"Token: {token.text}, POS: {token.pos_}, Tag: {token.tag_}")
````

## 8. Using NLTK for Stopword Removal

````python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Input text
words = ["This", "is", "a", "simple", "example"]

# Filter stopwords
filtered_words = [word for word in words if word.lower() not in stop_words]
print("Filtered Words:", filtered_words)
````

## 9. Using Hugging Face for Text Summarization

````python
from transformers import pipeline

# Load the summarization pipeline with a specific model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input text
text = """The 3MTT DeepTech Ready Program, spearheaded by Data Science Nigeria with support from Google,
is an innovative initiative designed to empower individuals with cutting-edge skills in technology. This
program focuses on fostering expertise in areas such as artificial intelligence, machine learning, and data
analytics, preparing participants to excel in the rapidly evolving tech landscape. Through hands-on training,
mentorship from industry experts, and real-world project implementation, the program aims to bridge the gap between
theoretical knowledge and practical application.
By equipping learners with these advanced skills, the 3MTT DeepTech Ready Program is contributing to building a robust
and competitive tech talent pool in Nigeria."""

# Generate summary
summary = summarizer(text, max_length=40, min_length=20, do_sample=False)
print("Summary:")
print(summary[0]['summary_text'])
````

## 10. Using spaCy for Sentence Similarity

````python
# Comparing sentence similarity with spaCy
doc1 = nlp("I like cats.")
doc2 = nlp("I enjoy dogs.")
similarity = doc1.similarity(doc2)
print(f"Similarity between sentences: {similarity:.2f}")
````

## 11. Using NLTK for Word Frequency Analysis

````python
from nltk.probability import FreqDist

# Input text
text = "Natural Language Processing enables machines to understand human language."

# Tokenize and analyze frequency
tokens = nltk.word_tokenize(text.lower())
freq_dist = FreqDist(tokens)

# Display frequencies
print("Word Frequencies:")
for word, frequency in freq_dist.most_common():
    print(f"{word}: {frequency}")
````

````python
%%markdown
## 12. Topic Modeling with Latent Dirichlet Allocation (LDA)

Topic Modeling is a technique used to discover the abstract "topics" that occur in a collection of documents. Latent Dirichlet Allocation (LDA) is a popular generative statistical model used for topic modeling. It assumes that each document is a mixture of a small number of topics and that each word's presence in a document is attributable to one of the document's topics. This section demonstrates how to perform LDA using scikit-learn.
````

**Reasoning**:
Implement LDA topic modeling using scikit-learn with sample data and display the results.



````python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Sample documents
documents = [
    "This is a document about machine learning and artificial intelligence.",
    "This document is about natural language processing and its applications.",
    "This document discusses deep learning and neural networks.",
    "This is another document about artificial intelligence and deep learning.",
    "This document is about computer vision and image processing."
]

# Create a CountVectorizer to convert documents to a matrix of token counts
vectorizer = CountVectorizer(stop_words='english')
data_vectorized = vectorizer.fit_transform(documents)

# Apply Latent Dirichlet Allocation (LDA)
# n_components is the number of topics to discover
lda_model = LatentDirichletAllocation(n_components=2, random_state=42)
lda_model.fit(data_vectorized)

# Display the top words for each topic
print("Topics and their top words:")
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda_model.components_):
    print(f"Topic #{topic_idx + 1}:")
    # Get the indices of the top words for the current topic
    top_words_indices = topic.argsort()[:-10 - 1:-1]
    # Get the actual words using the indices
    top_words = [feature_names[i] for i in top_words_indices]
    print(", ".join(top_words))

# Display the topic distribution for each document
print("\nTopic distribution for each document:")
topic_distributions = lda_model.transform(data_vectorized)
for i, doc_distribution in enumerate(topic_distributions):
    print(f"Document #{i + 1}: {doc_distribution}")
````
